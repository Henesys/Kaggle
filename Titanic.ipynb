```python
# Feature Engineering Approach
# https://triangleinequality.wordpress.com/2013/09/08/basic-feature-engineering-with-the-titanic-data/

def feature(df):

    # Title from "Name"
    df["Title"] = df["Name"].str.extract(" ([A-Za-z]+)\.", expand=False)
    
    # Size of Family (Including Passenger)
    # Some children travelled only with a nanny, therefore parch=0 for them.
    df["FamilySize"] = df["SibSp"] + df["Parch"] + 1
    
    # Letter of Cabin Location
    df["CabinLetter"] = df["Cabin"].str[0]

    # ValueError: Found unknown categories ['Countess', 'Ms'] in column 3 during transform
    rare_titles = ["Dona", "Lady", "Countess","Capt", "Col", "Don", 
                   "Dr", "Major", "Rev", "Sir", "Jonkheer", "Dona"]
    
    df["Title"] = df["Title"].replace(rare_titles, "Rare")
    df["Title"] = df["Title"].replace("Mlle", "Miss")
    df["Title"] = df["Title"].replace("Ms", "Miss")
    df["Title"] = df["Title"].replace("Mme", "Mrs")

    return df

feature_test = feature(test_data.copy())
feature_train = feature(train_data.copy())
```

```python
# Ensemble Learning Approach (Random Forest + Gradient Boost + Logistic Regression + SVM)

from sklearn.model_selection import train_test_split

from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import VotingClassifier

from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

# You have categorical data, but your model needs something numerical. See our one hot encoding tutorial for a solution.
# Need to convert categorical data via encoding
modified_features = ["Pclass", "Sex", "Age", "SibSp", "Parch", "Fare", "Embarked", "Title", "FamilySize", "CabinLetter"]
cat_features = ["Pclass", "Sex", "Embarked", "Title", "CabinLetter"]
num_features = ["Age", "SibSp", "Parch", "Fare", "FamilySize"]

X = feature_train[modified_features]
y = feature_train["Survived"]

# X = feature_train[modified_features]
# y = feature_train["Survived"]

# https://scikit-learn.org/stable/modules/generated/sklearn.compose.ColumnTransformer.html
# https://www.geeksforgeeks.org/using-columntransformer-in-scikit-learn-for-data-preprocessing/
preprocessor = ColumnTransformer(
    transformers=[
        ("num", Pipeline([
            ("imputer", SimpleImputer(strategy="median")),
            ("scaler", StandardScaler())
        ]), num_features),
        ("cat", Pipeline([
            ("imputer", SimpleImputer(strategy="most_frequent")),
            ("onehot", OneHotEncoder(drop="first", sparse_output=False, handle_unknown="ignore"))
        ]), cat_features)
    ])

# Ensemble Ingredients
rf = RandomForestClassifier(n_estimators=100, random_state=1984)
gb = GradientBoostingClassifier(n_estimators=100, random_state=1984)
lr = LogisticRegression(random_state=1984)
svm = SVC(probability=True, random_state=1984)

# Voting Classifier
# https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.VotingClassifier.html
# https://www.kaggle.com/code/marcinrutecki/voting-classifier-for-better-results
vote_pipeline = Pipeline([
    ("preprocessor", preprocessor),
    ("classifier", VotingClassifier(
        estimators=[("rf", rf), ("gb", gb), ("lr", lr), ("svm", svm)],
        voting="soft"
    ))
])

# Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1984)


# Fit the ensemble model
vote_pipeline.fit(X_train, y_train)

# Score 
train_score = vote_pipeline.score(X_train, y_train)
test_score = vote_pipeline.score(X_test, y_test)

print(f"Train score: {train_score:.5f}")
print(f"Test score: {test_score:.5f}")
```

```python
X_test_complete = feature_test[modified_features]
predictions = vote_pipeline.predict(X_test_complete)
```

```python
output = pd.DataFrame({'PassengerId': feature_test.PassengerId, 'Survived': predictions})
output.to_csv('submission.csv', index=False)
print("Your Ensemble Learning (w/ Feature Engineering Dataset) submission was successfully saved!")
```
